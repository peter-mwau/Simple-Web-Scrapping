```python
from bs4 import BeautifulSoup # For HTML parsing
import requests # Website connections
```

```python
url = ('https://commons.wikimedia.org/wiki/Commons:Village_pump') # URL to scrape

page = requests.get(url) # Connect to the URL
soup = BeautifulSoup(page.text, 'html') # Get the HTML
```

```python
soup # Print the HTML
```

```python
table = soup.find_all('table')[2] # Grab the third table
table
```

```python
table_titles = table.find_all('th') # Grab the table titles
table_titles
```

```python
table_titles = [title.strip() for title in table_titles] # Remove the HTML
table_titles
```

```python
import pandas as pd # For converting list to data frame
```

```python
df = pd.DataFrame(columns=table_titles) # Create a data frame to store the data
df
```

```python
column_data = table.find_all('tr') # Grab the table data
column_data
```

```python
for row in column_data[1:]: # Skip the first row
    row_data = row.find_all('td') # Grab the table data
    row_data = [data.text.strip() for data in row_data] # Remove the HTML
    print(row_data)
    length = len(df) # Get the length of the data frame
    df.loc[length] = row_data # Append data to the data frame
    
```
